```python
#Cálculo de tensão com variação no ângulo de aplicação da fibra

tau_xy = Stress_xy [0][0][0]
sigma_x = Stress_x [0][0][0] 
sigma_y = Stress_y [0][0][0]
theta = np.arange (0,91,1)
theta_radians = np.radians (theta)

x_local = []
y_local = []
tau_local = []

for i in range (len(theta)):
    a_i = cos(theta_radians[i])**2
    b_i = sin(theta_radians[i])**2
    c_i = 2*sin(theta_radians[i])*cos(theta_radians[i])
    d_i = sin(theta_radians[i])**2
    e_i = cos(theta_radians[i])**2
    f_i = -2*sin(theta_radians[i])*cos(theta_radians[i])
    g_i = -sin(theta_radians[i])*cos(theta_radians[i])
    h_i = sin(theta_radians[i])*cos(theta_radians[i])
    k_i = cos(theta_radians[i])**2-sin(theta_radians[i])**2
    xlocal = (a_i*sigma_x) + (b_i*sigma_y) + (c_i*tau_xy)
    ylocal = (d_i*sigma_x) + (e_i*sigma_y) + (f_i*tau_xy)
    taulocal = (g_i*sigma_x) + (h_i*sigma_y) + (k_i*tau_xy)
    x_local.append(xlocal)
    y_local.append(ylocal)
    tau_local.append(taulocal)
    
print ('---------------------------------------------------------------------------------------------------------------')
print ('Todos os resultados apresentados são com base na variação do ângulo Theta:')
print ('---------------------------------------------------------------------------------------------------------------') 
print('| Ângulo [°]    | Tensão em x [MPa]  | Tensão em y [MPa]  | Tensão de cisalhamento [MPa] |')
for j in range(len(theta)):
    print('|', ' '*(12-len(str('{:.2f}'.format(theta[j])))), '{:.2f}'.format(theta[j]), '|',
    ' '*(17-len(str('{:.3f}'.format(x_local[j])))), '{:.3f}'.format(x_local[j]), '|',
    ' '*(17-len(str('{:.3f}'.format(y_local[j])))), '{:.3f}'.format(y_local[j]), '|',
    ' '*(27-len(str('{:.3f}'.format(tau_local [j])))), '{:.3f}'.format(tau_local [j]), '|')
    print('')
print ('---------------------------------------------------------------------------------------------------------------')
print ('O valor de tensão mínima no eixo X será de:', np.round (np.min (x_local),4))
print ('O valor de tensão mínima no eixo Y será de:', np.round (np.min (y_local),4))
print ('O valor de tensão de cisalhamento mínima no plano XY será de:', np.round (np.min (tau_local),4))
```
